A simple linear, positive and conservative non-dimensional model would be the following：

$ d_t c_1 = c_2 - ac_1 , \;\; d_t c_2 = ac_1 -c_2, $

with the non-dimensional constant $a \geq 0 $ and initial values $c_1(0) = c^0_1 > 0 $ and $c_2(0) = c^0_2 > 0 $. Here,

$ p_{1,2} = c_2, \;\; d_{1,2} = ac_1,\;\;  p_{2,1} = ac_1, \;\; d_{2,1} = c_2,\;\;  p_{1,1} = p_{2,2} = d_{1,1} = p_{2,2} = 0.$

The model system describes exchange of mass between the constituents $c_1$ and $c_2$. The analytical solution is

$ c_1 =  \left \{ 1 + cexp[-(a + 1)t] \right \}c_1^\infty $

with the asymptotic solution

$ c_1^\infty = \frac{c_1^0+c_2^0}{a+1} \;\; and \;\; c=\frac{c_1^0}{c_1^\infty}-1.$

```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a       = 5
c_1_0   = 0.9
c_2_0   = 0.1
c_1_inf = (c_1_0 + c_2_0) / (a + 1)
c       = c_1_0 / c_1_inf -1
t       = np.linspace(0,1.8,181)
c_1     = (1 + c * np.exp(-(a + 1) * t)) * c_1_inf
c_2     = c_1_0 + c_2_0 - c_1

plt.plot(t, c_1, label='c1, analytical')
plt.plot(t, c_2, label='c2, analytical')
plt.plot(t, c_1+c_2, label='c1+c2, analytical')
plt.xlabel('time')
plt.ylabel('Concentration')
plt.title("Simple linear model")
plt.legend()
plt.grid(True)
plt.xlim(0, 1.8)
plt.ylim(0, 1.2)
plt.show()
```

```python
def rhs(c,a):
    rhs = np.zeros(c.size)
    p = np.zeros((c.size,c.size))
    d = np.zeros((c.size,c.size))
    p[0,1] = c[1]
    d[0,1] = a * c[0]
    p[1,0] = a * c[0]
    d[1,0] = c[1]
    for i in np.arange(c.size):
        rhs[i] = p[i,:].sum() - d[i,:].sum()
    return rhs
```

```python
def func_1(c,a,dt):
    dcdt = c + dt * rhs(c,a)
    return dcdt

def euler_forward(func,c_in,dt,tmax):
# First-order Euler-forward (E1) scheme.
# To evaluate the right-hand sides per time step.
# c_i^{n+1} = c_i^n  + \Delta t \left\{P_i(\underline{c}^n) - D_i(\underline{c}^n) \right\}.
# func: calculated value of the right-side hand terms.
# c_in: initial values of c_i^{0}, a vector, c_i^{0} = [c_1^{0}, c_2^{0}, ..., c_i^{0}].
# dt  : time step.
# t   : time scalar.
    row = math.ceil(tmax/dt)
    col = c_in.size
    c_n = np.zeros(row*col).reshape(row,col)
    c_n[0,:] = c_in
    for t in np.arange(row-1):
        c_n[t+1,:] = func(c_n[t,:],a,dt)
    return c_n
```

```python
def func_2(c,a,dt):
    dcdt = np.zeros(c.size)
    p = np.zeros((c.size,c.size))
    d = np.zeros((c.size,c.size))
    p[0,1] = c[1]
    d[0,1] = a * c[0]
    p[1,0] = a * c[0]
    d[1,0] = c[1]
    for i in np.arange(c.size):
        dcdt[i] = (c[i] + dt * p[i,:].sum())/(1 + dt * d[i,:].sum() / c[i])    
    return dcdt

def patankar(func,c_in,dt,tmax):
# First-order Patankar-Euler scheme (PE1) scheme.
# To evaluate the right-hand sides per time step.
# c_i^{n+1} = c_i^n  + \Delta t \left\{P_i(\underline{c}^n)-D_i(\underline{c}^n)\frac{c_i^{n+1}}{c_i^n} \right\}.
# func: calculated value of the right-side hand terms.
# c_in: initial values of c_i^{0}, a vector, c_i^{0} = [c_1^{0}, c_2^{0}, ..., c_i^{0}].
# dt  : time step.
# t   : time scalar.
    row = math.ceil(tmax/dt)
    col = c_in.size
    c_n = np.zeros(row*col).reshape(row,col)
    c_n[0,:] = c_in
    for t in np.arange(row-1):
        c_n[t+1,:] = func(c_n[t,:],a,dt)
    return c_n
```

```python
def func_3(c,a,dt):
    A = np.zeros((c.size,c.size))
    p = np.zeros((c.size,c.size))
    d = np.zeros((c.size,c.size))
    p[0,1] = c[1]
    d[0,1] = a * c[0]
    p[1,0] = a * c[0]
    d[1,0] = c[1]
    for i in np.arange(c.size):
        for j in np.arange(c.size):
            if i == j:
                A[i,j] = 1 + dt * d[i,:].sum() / c[i]
            else:
                A[i,j] = -dt * p[i,j] / c[j]
    return np.linalg.inv(A)

def modified_patankar(func,c_in,dt,tmax):
# First-order Modified Patankar-Euler scheme (MPE1) scheme.
# To evaluate the right-hand sides per time step.
# c_i^{n+1} = c_i^n+ \Delta t \left\{ \sum\limits_{\stackrel{j=1}{j \not= i}}^Ip_{i,j}(\underline{c}^n) 
# \dfrac{c_j^{n+1}}{c_j^n} - \sum_{j=1}^I d_{i,j}(\underline{c}^n)\dfrac{c_i^{n+1}}{c_i^n} \right\}.
# func: calculated value of the right-side hand terms.
# c_in: initial values of c_i^{0}, a vector, c_i^{0} = [c_1^{0}, c_2^{0}, ..., c_i^{0}].
# dt  : time step.
# t   : time scalar.
    row = math.ceil(tmax/dt)
    col = c_in.size
    c_n = np.zeros(row*col).reshape(row,col)
    c_n[0,:] = c_in
    for t in np.arange(row-1):
        c_n[t+1,:] = np.dot(func(c_n[t,:],a,dt), c_n[t,:].T).T
    return c_n
```

```python
def func_4(c,a,dt):
    c_1  = c + dt * rhs(c,a)
    dcdt = c + 0.5 * dt * (rhs(c,a) + rhs(c_1,a))
    return dcdt

def runge_kutta_2(func,c_in,dt,tmax):
# The second-order Runge-Kutta (RK2) scheme.
# To evaluate the right-hand sides per time step.
# c_i^{(1)} = c_i^n  + \Delta t \left\{P_i(\underline{c}^n) - D_i(\underline{c}^n) \right\}, \\ \\
# c_i^{n+1} = c_i^n  + \dfrac{\Delta t}{2}\left\{P_i(\underline{c}^n) + P_i(\underline{c}^{(1)}) 
# - D_i(\underline{c}^n) - D_i(\underline{c}^{(1)})\right\}.
# func: calculated value of the right-side hand terms.
# c_in: initial values of c_i^{0}, a vector, c_i^{0} = [c_1^{0}, c_2^{0}, ..., c_i^{0}].
# dt  : time step.
# t   : time scalar.
    row = math.ceil(tmax/dt)
    col = c_in.size
    c_n = np.zeros(row*col).reshape(row,col)
    c_n[0,:] = c_in
    for t in np.arange(row-1):
        c_n[t+1,:] = func(c_n[t,:],a,dt)
    return c_n
```

```python
def func_5(c,a,dt):
    dcdt = np.zeros(c.size)
    c_1  = func_2(c,a,dt)
    p_1 = np.zeros((c_1.size,c.size))
    d_1 = np.zeros((c_1.size,c.size))
    p_1[0,1] = c_1[1]
    d_1[0,1] = a * c_1[0]
    p_1[1,0] = a * c_1[0]
    d_1[1,0] = c_1[1]
    p = np.zeros((c.size,c.size))
    d = np.zeros((c.size,c.size))
    p[0,1] = c[1]
    d[0,1] = a * c[0]
    p[1,0] = a * c[0]
    d[1,0] = c[1]
    for i in np.arange(c.size):
        dcdt[i] = (c[i] + 0.5 * dt * (p[i,:].sum() + p_1[i,:].sum()))/(1 + 0.5 * dt * (d[i,:].sum() + d_1[i,:].sum()) / c_1[i]) 
    return dcdt

def patankar_runge_kutta_2(func,c_in,dt,tmax):
# The second-order Patankar-Runge-Kutta (PRK2) scheme.
# To evaluate the right-hand sides per time step.
# c_i^{(1)} = c_i^n  + \Delta t\left\{P_i(\underline{c}^n)- D_i(\underline{c}^n)\dfrac{c_i^{(1)}}{c_i^n}\right\},\\ \\
# c_i^{n+1} = c_i^n  + \dfrac{\Delta t}{2}\left\{P_i(\underline{c}^n)+ P_i(\underline{c}^{(1)})
# - ( D_i(\underline{c}^n)+ D_i(\underline{c}^{(1)}))\dfrac{c_i^{n+1}}{c_i^{(1)}}\right\}.
# func: calculated value of the right-side hand terms.
# c_in: initial values of c_i^{0}, a vector, c_i^{0} = [c_1^{0}, c_2^{0}, ..., c_i^{0}].
# dt  : time step.
# t   : time scalar.
    row = math.ceil(tmax/dt)
    col = c_in.size
    c_n = np.zeros(row*col).reshape(row,col)
    c_n[0,:] = c_in
    for t in np.arange(row-1):
        c_n[t+1,:] = func(c_n[t,:],a,dt)
    return c_n
```

```python
def func_6(c,a,dt):
    AA = np.zeros((c.size,c.size))
    cr3 = modified_patankar(func_3,c,dt,1)
    c_1 = cr3[1,:]
    p_1 = np.zeros((c_1.size,c.size))
    d_1 = np.zeros((c_1.size,c.size))
    p_1[0,1] = c_1[1]
    d_1[0,1] = a * c_1[0]
    p_1[1,0] = a * c_1[0]
    d_1[1,0] = c_1[1]
    p = np.zeros((c.size,c.size))
    d = np.zeros((c.size,c.size))
    p[0,1] = c[1]
    d[0,1] = a * c[0]
    p[1,0] = a * c[0]
    d[1,0] = c[1]
    pp = 0.5 * (p + p_1)
    dd = 0.5 * (d + d_1)
    for i in np.arange(c.size):
        for j in np.arange(c.size):
            if i == j:
                AA[i,j] = 1 + dt * dd[i,:].sum() / c_1[i]
            else:
                AA[i,j] = -dt * pp[i,j] / c_1[j]
    return np.linalg.inv(AA)

def modified_patankar_runge_kutta_2(func,c_in,dt,tmax):
# Modified Patankar-Runge-Kutta (MPRK) scheme.
# To evaluate the right-hand sides per time step.
# func: calculated value of the right-side hand terms.
# c_in: initial values of c_i^{0}, a vector, c_i^{0} = [c_1^{0}, c_2^{0}, ..., c_i^{0}].
# dt  : time step.
# t   : time scalar.
    row = math.ceil(tmax/dt)
    col = c_in.size
    c_n = np.zeros(row*col).reshape(row,col)
    c_n[0,:] = c_in
    for t in np.arange(row-1):
        c_n[t+1,:] = np.dot(func(c_n[t,:],a,dt), c_n[t,:].T).T
    return c_n
```

```python
def func_7(c,cc,a,dt,tmax):
    c_r_1 = c + dt * rhs(c,a)
    c_r_n1 = c + 0.5 * dt * (rhs(c,a) + rhs(c_r_1,a))
    if min(min(c_r_1),min(c_r_n1))>0:
        dcdt = c_r_n1
    else:
        dcdt = cc
    return dcdt

def hybrid_modified_patankar_runge_kutta_2(func,c_in,dt,tmax):
# Hybrid modified Patankar–Runge–Kutta scheme (HMPRK).
# To evaluate the right-hand sides per time step.
# func: calculated value of the right-side hand terms.
# c_in: initial values of c_i^{0}, a vector, c_i^{0} = [c_1^{0}, c_2^{0}, ..., c_i^{0}].
# dt  : time step.
# t   : time scalar.
    row = math.ceil(tmax/dt)
    col = c_in.size
    c_n = np.zeros(row*col).reshape(row,col)
    c_n[0,:] = c_in
    cr6 = modified_patankar_runge_kutta_2(func_6,c_in,dt,tmax)
    for t in np.arange(row-1):
        c_n[t+1,:] = func(c_n[t,:],cr6[t+1,:],a,dt,tmax)
    return c_n
```

```python
a = 5
dt = 0.25
tmax = 1.8
c_in = np.array([0.9,0.1])

cr1 = euler_forward(func_1,c_in,dt,tmax)
cr2 = patankar(func_2,c_in,dt,tmax)
cr3 = modified_patankar(func_3,c_in,dt,tmax)
cr4 = runge_kutta_2(func_4,c_in,dt,tmax)
cr5 = patankar_runge_kutta_2(func_5,c_in,dt,tmax)
cr6 = modified_patankar_runge_kutta_2(func_6,c_in,dt,tmax)
cr7 = hybrid_modified_patankar_runge_kutta_2(func_7,c_in,dt,tmax)
```

```python
t_s = np.arange(0,tmax,dt)

golden_ratio = (1.+np.sqrt(5.))/2.
figwidth     = 7.48
figsize = (figwidth,figwidth/golden_ratio*4/2) 
fig, axs = plt.subplots(4, 2, sharex='col', sharey='all', figsize=figsize)

axs[0, 0].plot(t, c_1, '-', label='c1, analytical')
axs[0, 0].plot(t, c_2, '-', label='c2, analytical')
axs[0, 0].plot(t, c_1 + c_2, '-', label='c1+c2, analytical')
axs[0, 0].set(title='Simple linear model', ylabel='Concentration')

axs[1, 0].plot(t, c_1, '-', label='c1, analytical')
axs[1, 0].plot(t, c_2, '-', label='c2, analytical')
axs[1, 0].plot(t, c_1 + c_2, '-', label='c1+c2, analytical')
axs[1, 0].plot(t_s, cr1[:,0], 'o--', label='c1, simulated')
axs[1, 0].plot(t_s, cr1[:,1], 'o--', label='c2, simulated', markerfacecolor='none')
axs[1, 0].plot(t_s, cr1[:,0] + cr1[:,1], 's--', label='c2, simulated')
axs[1, 0].set(title='Forward Euler scheme (E)', ylabel='Concentration')

axs[2, 0].plot(t, c_1, '-', label='c1, analytical')
axs[2, 0].plot(t, c_2, '-', label='c2, analytical')
axs[2, 0].plot(t, c_1 + c_2, '-', label='c1+c2, analytical')
axs[2, 0].plot(t_s, cr2[:,0], 'o--', label='c1, simulated')
axs[2, 0].plot(t_s, cr2[:,1], 'o--', label='c2, simulated', markerfacecolor='none')
axs[2, 0].plot(t_s, cr2[:,0] + cr2[:,1], 's--', label='c2, simulated')
axs[2, 0].set(title='Patankar scheme (PE)', ylabel='Concentration')

axs[3, 0].plot(t, c_1, '-', label='c1, analytical')
axs[3, 0].plot(t, c_2, '-', label='c2, analytical')
axs[3, 0].plot(t, c_1 + c_2, '-', label='c1+c2, analytical')
axs[3, 0].plot(t_s, cr3[:,0], 'o--', label='c1, simulated')
axs[3, 0].plot(t_s, cr3[:,1], 'o--', label='c2, simulated', markerfacecolor='none')
axs[3, 0].plot(t_s, cr3[:,0] + cr3[:,1], 's--', label='c2, simulated')
axs[3, 0].set(title='Modified Patankar scheme (MPE)', xlabel='time', ylabel='Concentration')

axs[0, 1].plot(t, c_1, '-', label='c1, analytical')
axs[0, 1].plot(t, c_2, '-', label='c2, analytical')
axs[0, 1].plot(t, c_1 + c_2, '-', label='c1+c2, analytical')
axs[0, 1].plot(t_s, cr4[:,0], 'o--', label='c1, simulated')
axs[0, 1].plot(t_s, cr4[:,1], 'o--', label='c2, simulated', markerfacecolor='none')
axs[0, 1].plot(t_s, cr4[:,0] + cr4[:,1], 's--', label='c2, simulated')
axs[0, 1].set(title='Runge–Kutta scheme (RK)')

axs[1, 1].plot(t, c_1, '-', label='c1, analytical')
axs[1, 1].plot(t, c_2, '-', label='c2, analytical')
axs[1, 1].plot(t, c_1 + c_2, '-', label='c1+c2, analytical')
axs[1, 1].plot(t_s, cr5[:,0], 'o--', label='c1, simulated')
axs[1, 1].plot(t_s, cr5[:,1], 'o--', label='c2, simulated', markerfacecolor='none')
axs[1, 1].plot(t_s, cr5[:,0] + cr5[:,1], 's--', label='c2, simulated')
axs[1, 1].set(title='Patankar–Runge–Kutta scheme (PRK)')

axs[2, 1].plot(t, c_1, '-', label='c1, analytical')
axs[2, 1].plot(t, c_2, '-', label='c2, analytical')
axs[2, 1].plot(t, c_1 + c_2, '-', label='c1+c2, analytical')
axs[2, 1].plot(t_s, cr6[:,0], 'o--', label='c1, simulated')
axs[2, 1].plot(t_s, cr6[:,1], 'o--', label='c2, simulated', markerfacecolor='none')
axs[2, 1].plot(t_s, cr6[:,0] + cr6[:,1], 's--', label='c2, simulated')
axs[2, 1].set(title='Modified PRK scheme (MPRK)')

axs[3, 1].plot(t, c_1, '-', label='c1, analytical')
axs[3, 1].plot(t, c_2, '-', label='c2, analytical')
axs[3, 1].plot(t, c_1 + c_2, '-', label='c1+c2, analytical')
axs[3, 1].plot(t_s, cr7[:,0], 'o--', label='c1, simulated')
axs[3, 1].plot(t_s, cr7[:,1], 'o--', label='c2, simulated', markerfacecolor='none')
axs[3, 1].plot(t_s, cr7[:,0] + cr7[:,1], 's--', label='c2, simulated')
axs[3, 1].set(title='Hybrid MPRK scheme (HMPRK)')

plt.show()
```
